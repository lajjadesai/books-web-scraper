```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/books_detailed.csv", encoding="utf-8", encoding_errors="ignore")
df.head()
```

```python
# Clean and convert price to float
df["price"] = df["price"].astype(str).str.replace(r"[^\d.]", "", regex=True).astype(float)

# Convert rating to number
rating_map = {
    "One": 1, "Two": 2, "Three": 3,
    "Four": 4, "Five": 5
}
df["rating_num"] = df["rating"].map(rating_map)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(df["price"], bins=20, kde=True)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Number of Books")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="rating_num", data=df)
plt.title("Book Ratings Distribution")
plt.xlabel("Rating (1 to 5)")
plt.ylabel("Number of Books")
plt.show()
```

```python
top_categories = df["category"].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top 10 Book Categories")
plt.xlabel("Number of Books")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(x="rating_num", y="price", data=df)
plt.title("Average Book Price by Rating")
plt.xlabel("Rating")
plt.ylabel("Avg Price (£)")
plt.show()
```
